#Question 2: Student Final Grade Calculator (Global & Local Variables)

**Description**

Create a student grade calculator that uses both global and local variables to calculate a student's
final grade based on multiple assessment components.

**Requirements**

1. Use a global variable to store the total maximum marks (e.g., 500)
2. Create a function that takes individual subject marks as local variables
3. Calculate the percentage using both global and local variables
4. Determine the grade based on percentage:
- A: 90% and above
- B: 80-89%
- C: 70-79%
- D: 60-69%
- F: Below 60%
5. Display student name, total marks, percentage, and final grade

**Input/Output Example**

Enter student name: John Doe

Enter marks for Math: 85

Enter marks for Science: 90

Enter marks for English: 88

Enter marks for History: 82

Enter marks for Geography: 90

Student: John Doe

Total Marks Obtained: 435

Maximum Marks: 500

Percentage: 87.0%

Final Grade: B

**Constraints**

- Each subject mark should be between 0 and 100
- Validate all inputs

```python
# Global variable
MAX_MARKS = 500

def calculate_grade(math, science, english, history, geography):
    # Local variables
    total_marks = math + science + english + history + geography
    percentage = (total_marks / MAX_MARKS) * 100

    if percentage >= 90:
        grade = "A"
    elif percentage >= 80:
        grade = "B"
    elif percentage >= 70:
        grade = "C"
    elif percentage >= 60:
        grade = "D"
    else:
        grade = "F"

    return total_marks, percentage, grade


def get_valid_marks(subject):
    while True:
        try:
            marks = float(input(f"Enter marks for {subject}: "))

            if 0 <= marks <= 100:
                return marks
            else:
                print("Marks must be between 0 and 100.")

        except ValueError:
            print("Invalid input! Please enter a number.")

student_name = input("Enter student name: ")

math = get_valid_marks("Math")
science = get_valid_marks("Science")
english = get_valid_marks("English")
history = get_valid_marks("History")
geography = get_valid_marks("Geography")

total, percentage, grade = calculate_grade(
    math, science, english, history, geography
)

print("\n----- STUDENT REPORT -----")
print("Student:", student_name)
print("Total Marks Obtained:", total)
print("Maximum Marks:", MAX_MARKS)
print(f"Percentage: {percentage:.2f}%")
print("Final Grade:", grade)
```

```python
CORRECT_PASSWORD = "SecurePass123!"
MAX_ATTEMPTS = 3
attempts = 0
print("=== Password Validation System ===")

while attempts < MAX_ATTEMPTS:
    password = input("Enter password: ")
    if password == "":
        remaining = MAX_ATTEMPTS - attempts
        print(f"Password cannot be empty. {remaining} attempts remaining.")
        continue
    if password == CORRECT_PASSWORD:
        print("Access granted! Welcome.")
        break
    attempts += 1 #incorrect
    remaining = MAX_ATTEMPTS - attempts

    if remaining > 0:
        print(f"Incorrect password. {remaining} attempt(s) remaining.")
    else:
        print("Maximum attempts exceeded. Account locked.")
```

#Question 7: Data Processing with Loops (ETL Simulation)

**Description**

Simulate an ETL (Extract, Transform, Load) process using loops to process transaction data.

**Requirements**

1. Create a list of transaction dictionaries with the following structure:

python { "transaction_id": 1, "product_id": 101, "amount": 200, "date":
"2025-01-10" }
2. Use a for loop to iterate through transactions
3. Skip transactions with missing or invalid data (use continue)
4. Calculate total sales per product
5. Count valid and invalid transactions
6. Display summary report

**Input/Output Example**

=== ETL Data Processing ===

Processing transactions...

Skipping transaction 3 due to missing amount.

=== Summary Report ===

Total Transactions: 5

Valid Transactions: 4

Invalid Transactions: 1

Product Sales:

Product ID 101: Rs 450

Product ID 102: Rs 300

Product ID 103: Rs 150

**Constraints**

- Handle None values and missing keys
- Use continue for invalid transactions
- Use dictionaries to aggregate data

```python
print("=== ETL Data Processing ===")
print("Processing transactions...\n")
transactions = [
    {"transaction_id": 1, "product_id": 101, "amount": 200, "date": "2025-01-10"},
    {"transaction_id": 2, "product_id": 102, "amount": 300, "date": "2025-01-11"},
    {"transaction_id": 3, "product_id": 101, "amount": None, "date": "2025-01-12"},  # Invalid
    {"transaction_id": 4, "product_id": 103, "amount": 150, "date": "2025-01-13"},
    {"transaction_id": 5, "product_id": 101, "amount": 250, "date": "2025-01-14"}
]

total_transactions = len(transactions)
valid_transactions = 0
invalid_transactions = 0
product_sales = {}

for transaction in transactions:
    required_keys = ["transaction_id", "product_id", "amount", "date"]

    if not all(key in transaction for key in required_keys):
        print(f"Skipping transaction due to missing key: {transaction}")
        invalid_transactions += 1
        continue

    # Check invalid amount
    if transaction["amount"] is None or transaction["amount"] <= 0:
        print(
            f"Skipping transaction {transaction['transaction_id']} due to invalid amount."
        )
        invalid_transactions += 1
        continue

    # Extract
    product_id = transaction["product_id"]
    amount = transaction["amount"]

    # Transform + Load (Aggregation)
    if product_id in product_sales:
        product_sales[product_id] += amount
    else:
        product_sales[product_id] = amount

    valid_transactions += 1

print("\n=== Summary Report ===")
print(f"Total Transactions: {total_transactions}")
print(f"Valid Transactions: {valid_transactions}")
print(f"Invalid Transactions: {invalid_transactions}")

print("\nProduct Sales:")
for product_id, total_sales in product_sales.items():
    print(f"Product ID {product_id}: Rs {total_sales}")
```

2. Perform the following operations:
- Count total characters (including spaces)
- Count words
- Count sentences
- Find most frequent word
- Extract all unique words (using set)
- Create a dictionary with word frequencies
- Reverse the entire text
- Extract first and last sentence

**Input/Output Example**

Enter text: Python is great. Python is easy. Python is powerful.

=== Text Analysis Report ===

Total Characters: 47

Total Words: 9

Total Sentences: 3

Most Frequent Word: Python (appears 3
times)

Unique Words: {'Python', 'is', 'great', 'easy', 'powerful'}

Word Frequencies: {'Python': 3, 'is': 3, 'great': 1, 'easy': 1, 'powerful': 1}

Reversed Text: .lufrewop si nohtyP .ysae si nohtyP .taerg si nohtyP

First Sentence: Python is great.

Last Sentence: Python is powerful.

**Constraints**
- Handle punctuation properly
- Case-insensitive word counting
- Use appropriate collection types

```python
text = input("Enter text: ").strip()
if not text:
    print("Error: Text cannot be empty.")
else:
    total_characters = len(text)
    sentences = [s.strip() for s in text.split('.') if s.strip()]
    total_sentences = len(sentences)

    cleaned_text = text

    for char in ".,!?;:":
        cleaned_text = cleaned_text.replace(char, "")
    words = cleaned_text.lower().split()
    total_words = len(words)

    word_frequencies = {}

    for word in words:
        if word in word_frequencies:
            word_frequencies[word] += 1
        else:
            word_frequencies[word] = 1

    most_frequent_word = max(
        word_frequencies,
        key=word_frequencies.get
    )
    unique_words = set(words)
    reversed_text = text[::-1]

    first_sentence = sentences[0] + "." if sentences else ""
    last_sentence = sentences[-1] + "." if sentences else ""

    print("\n=== Text Analysis Report ===")
    print("Total Characters:", total_characters)
    print("Total Words:", total_words)
    print("Total Sentences:", total_sentences)

    print(
        f"Most Frequent Word: {most_frequent_word} "
        f"(appears {word_frequencies[most_frequent_word]} times)"
    )

    print("Unique Words:", unique_words)

    print("Word Frequencies:", word_frequencies)

    print("Reversed Text:", reversed_text)

    print("First Sentence:", first_sentence)

    print("Last Sentence:", last_sentence)
```

```python
history = []

# Arithmetic Operations
def arithmetic_operations():
    try:
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))
        op = input("Select operation (+, -, *, /, //, %, **): ")
        if op == "+":
            result = num1 + num2
        elif op == "-":
            result = num1 - num2
        elif op == "*":
            result = num1 * num2
        elif op == "/":
            if num2 == 0:
                print("Error: Division by zero!")
                return
            result = num1 / num2
        elif op == "//":
            if num2 == 0:
                print("Error: Division by zero!")
                return
            result = num1 // num2
        elif op == "%":
            if num2 == 0:
                print("Error: Division by zero!")
                return
            result = num1 % num2
        elif op == "**":
            result = num1 ** num2
        else:
            print("Invalid operation!")
            return

        print("Result:", result)
        history.append(f"{num1} {op} {num2} = {result}")

    except ValueError:
        print("Invalid numeric input!")

# Comparison Operations
def comparison_operations():
    try:
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))

        print("\nComparison Results:")
        print(f"{num1} > {num2} : {num1 > num2}")
        print(f"{num1} < {num2} : {num1 < num2}")
        print(f"{num1} >= {num2} : {num1 >= num2}")
        print(f"{num1} <= {num2} : {num1 <= num2}")
        print(f"{num1} == {num2} : {num1 == num2}")
        print(f"{num1} != {num2} : {num1 != num2}")
        history.append(f"Compared {num1} and {num2}")
    except ValueError:
        print("Invalid numeric input!")

# Logical Operations
def logical_operations():
    try:
        age = int(input("Enter age: "))
        income = float(input("Enter income: "))

        eligible = age >= 18 and income >= 10000
        senior = age >= 60 or income >= 50000
        not_adult = not (age >= 18)

        print("\nLogical Results:")
        print("Eligible:", eligible)
        print("Senior Benefit:", senior)
        print("Not Adult:", not_adult)

        history.append(
            f"Logical Check -> Age:{age}, Income:{income}")
    except ValueError:
        print("Invalid input!")

# String Operations
def string_operations():
    str1 = input("Enter first string: ")
    str2 = input("Enter second string: ")

    op = input("Select operation (concat/repeat): ").lower()

    if op == "concat":
        result = str1 + str2

        print("Result:", result)

        history.append(
            f"{str1} + {str2} = {result}"
        )

    elif op == "repeat":
        try:
            times = int(input("Repeat how many times? "))

            result = str1 * times

            print("Result:", result)

            history.append(
                f"{str1} repeated {times} times = {result}"
            )

        except ValueError:
            print("Invalid repetition count!")

    else:
        print("Invalid operation!")

# Collection Operations
def collection_operations():

    numbers = []

    try:
        n = int(input("How many numbers to add to the list? "))

        for i in range(n):
            value = float(input(f"Enter number {i+1}: "))
            numbers.append(value)

        print("\nList:", numbers)

        if numbers:
            print("Maximum:", max(numbers))
            print("Minimum:", min(numbers))
            print("Sum:", sum(numbers))

            history.append(
                f"List Operation on {numbers}"
            )

    except ValueError:
        print("Invalid input!")

# View History
def view_history():

    print("\n=== Calculation History ===")

    if not history:
        print("No calculations performed yet.")
    else:
        for i, item in enumerate(history, start=1):
            print(f"{i}. {item}")

while True:

    print("\n=== Comprehensive Calculator ===")
    print("1. Arithmetic Operations")
    print("2. Comparison Operations")
    print("3. Logical Operations")
    print("4. String Operations")
    print("5. Collection Operations")
    print("6. View History")
    print("7. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        arithmetic_operations()

    elif choice == "2":
        comparison_operations()

    elif choice == "3":
        logical_operations()

    elif choice == "4":
        string_operations()

    elif choice == "5":
        collection_operations()

    elif choice == "6":
        view_history()

    elif choice == "7":
        print("Thank you for using the calculator!")
        break

    else:
        print("Invalid choice! Please try again.")
```
