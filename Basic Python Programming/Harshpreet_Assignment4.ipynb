#Question 1: Library Management System (Classes & Objects)

```python
class Book:
    def __init__(
        self,
        title,
        author,
        isbn,
        total_copies
    ):
        if not title.strip():
            raise ValueError(
                "Title cannot be empty."
            )

        if not author.strip():
            raise ValueError(
                "Author cannot be empty."
            )

        if total_copies <= 0:
            raise ValueError(
                "Total copies must be positive."
            )

        self.title = title
        self.author = author
        self.isbn = isbn
        self.total_copies = total_copies
        self.copies_available = total_copies

    def display_info(self):

        print(
            f"Title: {self.title}"
        )

        print(
            f"Author: {self.author}"
        )

        print(
            f"ISBN: {self.isbn}"
        )

        print(
            f"Available: "
            f"{self.copies_available}/"
            f"{self.total_copies}"
        )

        print("-" * 30)

    def borrow_book(self):

        if self.copies_available > 0:

            self.copies_available -= 1

            print(
                f"You have borrowed "
                f"'{self.title}'."
            )

            print(
                f"Copies left: "
                f"{self.copies_available}"
            )

        else:

            print(
                f"Sorry! "
                f"'{self.title}' "
                f"is not available."
            )

    def return_book(self):

        if (
            self.copies_available
            < self.total_copies
        ):

            self.copies_available += 1

            print(
                f"Book returned: "
                f"{self.title}"
            )

            print(
                f"Available copies: "
                f"{self.copies_available}"
            )

        else:

            print(
                "All copies are already "
                "in the library."
            )

    def is_available(self):

        return (
            self.copies_available > 0
        )

class Library:

    def __init__(self, name):

        self.name = name
        self.books = []

    def add_book(self, book):

        self.books.append(book)

        print(
            f"Book added: "
            f"{book.title} "
            f"by {book.author}"
        )

    def remove_book(self, title):

        for book in self.books:

            if (
                book.title.lower()
                ==
                title.lower()
            ):

                self.books.remove(book)

                print(
                    f"Book removed: "
                    f"{title}"
                )

                return

        print(
            f"Book '{title}' "
            f"not found."
        )

    def search_book(self, keyword):

        found = False

        for book in self.books:

            if (
                keyword.lower()
                in
                book.title.lower()
                or
                keyword.lower()
                in
                book.author.lower()
            ):

                print(
                    f"Found: "
                    f"{book.title} "
                    f"by "
                    f"{book.author}"
                )

                found = True

        if not found:

            print(
                "No matching books found."
            )

    def display_all_books(self):

        if not self.books:

            print(
                "Library is empty."
            )

            return

        print(
            "\n=== All Books ==="
        )

        for book in self.books:

            book.display_info()

    def borrow_book(self, title):

        for book in self.books:

            if (
                book.title.lower()
                ==
                title.lower()
            ):

                book.borrow_book()

                return

        print(
            f"Book '{title}' "
            f"not found."
        )

    def return_book(self, title):

        for book in self.books:

            if (
                book.title.lower()
                ==
                title.lower()
            ):

                book.return_book()

                return

        print(
            f"Book '{title}' "
            f"not found."
        )


print(
    "=== Library Management System ===\n"
)

library = Library(
    "Central Library"
)

# Create Book Objects

book1 = Book(
    "The Alchemist",
    "Paulo Coelho",
    "978-0061122415",
    3
)

book2 = Book(
    "Atomic Habits",
    "James Clear",
    "978-0735211292",
    2
)

book3 = Book(
    "Python Crash Course",
    "Eric Matthes",
    "978-1593279288",
    1
)

# Add Books

print(
    "Adding books to library..."
)

library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

# Display All Books

library.display_all_books()

# Borrow Book

print(
    "\nBorrowing "
    "'The Alchemist'..."
)

library.borrow_book(
    "The Alchemist"
)

# Borrow Again

library.borrow_book(
    "The Alchemist"
)

# Return Book

print(
    "\nReturning "
    "'The Alchemist'..."
)

library.return_book(
    "The Alchemist"
)

# Search Book

print(
    "\nSearching "
    "for 'Alchemist'..."
)

library.search_book(
    "Alchemist"
)

# Search Author

print(
    "\nSearching "
    "for 'James'..."
)

library.search_book(
    "James"
)

# Remove Book

print(
    "\nRemoving "
    "'Atomic Habits'..."
)

library.remove_book(
    "Atomic Habits"
)

# Display Remaining Books

library.display_all_books()
```

#Question 2: Banking System with Encapsulation

```python
class BankAccount:
  def __init__(self, account_number, account_holder, balance, pin):

    if not account_number.strip():
      raise ValueError("Account number cannot be empty.")

    if not account_holder.strip():
      raise ValueError("Account holder cannot be empty.")

    if balance < 0:
      raise ValueError("Initial balance cannot be negative.")

    if (not str(pin).isdigit()) or len(str(pin)) != 4:
      raise ValueError("Invalid PIN. It should be a 4-digit number.")

    self.__account_number = (account_number)
    self.__account_holder = (account_holder)
    self.__balance = (balance)
    self.__pin = (pin)

    print(f"Account created. \nAccount number : {account_number} \nAccount holder : {account_holder} \nBalance : {balance}")

  def get_balance(self, pin):
    if str(pin) == self.__pin:
      print(f"current balance is: {self.__balance}")
    print("Invalid pin. \nAccess denied.")
    return None

  def deposit(self,amount):
    if amount>0:
      self.__balance += amount
    else: raise ValueError("Deposit cannot be negative.")

  def withdraw(self,pin, amount):
    if pin != self.__pin:
      print("Invalid pin. \nAccess denied.")

    elif self.__balance >= amount:
      self.__balance -= amount

  def change_pin(self, old_pin, new_pin):
    if old_pin == self.__pin:
      self.__pin = new_pin
    else:
      print("Invalid pin. \nAccess denied.")

  def display_account_info(self, pin):
    if str(pin) == self.__pin:
      print(f"Account number : {self.__account_number} \nAccount holder : {self.__account_holder} \nBalance : {self.__balance}")

  print("=== Banking System ===\n")

print(
    "Creating account "
    "for John Doe..."
)

account = BankAccount(
    "ACC001",
    "John Doe",
    0,
    "1234"
)

print(
    "\nDepositing "
    "Rs 5000..."
)

account.deposit(
    5000
)

print(
    "\nChecking balance "
    "(PIN: 1234)..."
)

account.get_balance(
    "1234"
)

print(
    "\nWithdrawing "
    "Rs 2000 "
    "(PIN: 1234)..."
)

account.withdraw(
    2000,
    "1234"
)

print(
    "\nAttempting to "
    "access balance "
    "with wrong PIN..."
)

account.get_balance(
    "9999"
)

print(
    "\nAttempting to "
    "withdraw more "
    "than balance..."
)

account.withdraw(
    10000,
    "1234"
)

print(
    "\nChanging PIN..."
)

account.change_pin(
    "1234",
    "5678"
)

print(
    "\nDisplaying account "
    "information..."
)

account.display_account_info(
    "5678"
)

print(
    "\nAttempting direct "
    "access to balance..."
)

try:

    print(
        account.__balance
    )

except AttributeError:

    print(
        "Cannot access "
        "private attribute "
        "__balance directly."
    )
```

#Question 3: Employee Hierarchy with Inheritance

```python
class Employee:
  def __init__(self, name, employee_id, base_salary):
    if base_salary <0:
      raise ValueError("Base salary cannot be negative.")

    self.name = name
    self.employee_id = employee_id
    self.base_salary = base_salary

  def calculate_salary(self):
    return self.base_salary

  def display_info(self):
    print(f"Employee: {self.name} (ID: {self.employee_id})")
    print(f"Salary: Rs {self.calculate_salary()}")
    print(f"Role: {self.get_role()}")

  def get_role(self):
    return "Employee"
```

```python
class Manager(Employee):
  def __init__(self, name, employee_id, base_salary,department, bonus):

    super().__init__(name, employee_id, base_salary)
    self.department = department
    self.bonus = bonus

  def calculate_salary(self):
     return self.base_salary + self.bonus

  def get_role(self):
    return "Manager"

  def display_info(self):
    super().display_info()
    print(f"Bonus: {self.bonus}")
    print(f"Department: {self.department}")
```

```python
class Developer(Employee):
  def __init__(self, name, employee_id, base_salary, programming_language, overtime_hours):
     super().__init__(name, employee_id, base_salary)
     self.programming_language = programming_language
     self.overtime_hours = overtime_hours

  def calculate_salary(self):
    return self.base_salary + (self.overtime_hours * 500)

  def get_role(self):
     return "Developer"

  def display_info(self):
     super().display_info()
     print(f"Programming Language: {self.programming_language}")
     print(f"Overtime Hours: {self.overtime_hours}")
```

```python
print(
    "=== Employee Management System ===\n"
)

print(
    "Creating employees...\n"
)

employee = Employee(
    "John Doe",
    "E001",
    50000
)

manager = Manager(
    "Jane Smith",
    "M001",
    70000,
    "IT",
    10000
)

developer = Developer(
    "Bob Johnson",
    "D001",
    60000,
    "Python",
    10
)
employees = [
    employee,
    manager,
    developer
]

for emp in employees:

    emp.display_info()

    print("-" * 40)
```

#Question 4: Shape Calculator with Polymorphism

```python
import math

class Shape:
  def area(self):
    raise NotImplementedError("Subclasses must implement area().")
  def perimeter(self):
    raise NotImplementedError("Subclasses must implement perimeter().")
  def display_info(self):
    print(f"Area: {self.area(): .2f}")
    print(f"Perimeter: {self.perimeter(): .2f}")
```

```python
class Circle(Shape):
  def __init__(self, radius):
    if radius<0:
      raise ValueError("Radius cannot be negative")

    self.radius = radius
  def area(self):
    return math.pi * self.radius ** 2
  def perimeter(self):
    return 2 * math.pi * self.radius
  def display_info(self):
    print(f"Circle with radius {self.radius}:")
    super().display_info()
```

```python
class Rectangle(Shape):
  def __init__(self, length, width):
    if length<0 or width<0:
      raise ValueError("Length and width cannot be negative")

    self.length = length
    self.width = width
  def area(self):
    return self.length * self.width
  def perimeter(self):
    return 2 * (self.length + self.width)
  def display_info(self):
    print(f"Rectangle with length {self.length} and width {self.width}:")
    return super().display_info()
```

```python
class Triangle(Shape):
  def __init__(self, side1, side2, side3):
    if side1<=0 or side2<=0 or side3<=0:
      raise ValueError("Sides cannot be negative or zero")
    if side1 + side2 <= side3 or side1 + side3 <= side2 or side2 + side3 <= side1:
      raise ValueError("Invalid triangle sides")
    self.side1 = side1
    self.side2 = side2
    self.side3 = side3
  def area(self):
    s = (self.side1 + self.side2 + self.side3) / 2
    return math.sqrt(s * (s - self.side1) * (s - self.side2) * (s - self.side3))
  def perimeter(self):
    return self.side1 + self.side2 + self.side3
  def display_info(self):
    print(f"Triangle with sides {self.side1}, {self.side2}, and {self.side3}:")
    return super().display_info()
```

```python
print(
    "=== Shape Calculator ===\n"
)

print(
    "Creating shapes...\n"
)

circle = Circle(5)

rectangle = Rectangle(
    4,
    6
)

triangle = Triangle(
    3,
    4,
    5
)

circle.display_info()

print("-" * 35)

rectangle.display_info()

print("-" * 35)

triangle.display_info()

print("-" * 35)

print(
    "\nPolymorphic Behavior:"
)

print(
    "Processing list of shapes...\n"
)

shapes = [
    circle,
    rectangle,
    triangle
]

for index, shape in enumerate(
    shapes,
    start=1
):

    print(
        f"Shape {index}: "
        f"Area = "
        f"{shape.area():.2f}, "
        f"Perimeter = "
        f"{shape.perimeter():.2f}"
    )
```

#Question 5: Vehicle System with Multiple Inheritance

```python
class Engine:
  def __init__(self, fuel_type):
    self.fuel_type = fuel_type
  def start_engine(self):
    print(f"Engine start \nfuel type is: {self.fuel_type}")
  def stop_engine(self):
    print("Engine stop")
```

```python
class Electric:
  def __init__(self, battery_capacity):
     self.battery_capacity = battery_capacity
  def charge_battery(self):
    print(f"Battery charged \ncapacity is: {self.battery_capacity}")
```

```python
class Combustion:
  def __init__(self, fuel_tank_capacity):
    self.fuel_tank_capacity = fuel_tank_capacity
  def refuel(self):
    print(f"Fuel tank filled \ncapacity is: {self.fuel_tank_capacity}")
```

```python
class HybridCar(Engine, Electric, Combustion):
  def __init__(self, battery_capacity, fuel_tank_capacity):
    Engine.__init__(self, "Gasoline + Electric")
    Electric.__init__(self, battery_capacity)
    Combustion.__init__(self, fuel_tank_capacity)

  def vehicle_info(self):
    print("Hybrid Car")
    self.start_engine()
    self.charge_battery()
    self.refuel()
```

```python
class ElectricCar(Engine, Electric):
  def __init__(self, battery_capacity):
    Engine.__init__(self, "Electric")
    Electric.__init__(self, battery_capacity)

  def vehicle_info(self):
    print("Electric Car")
    self.start_engine()
    self.charge_battery()
```

```python
class GasCar(Engine, Combustion):
  def __init__(self, fuel_tank_capacity):
    Engine.__init__(self, "Gasoline")
    Combustion.__init__(self, fuel_tank_capacity)

  def vehicle_info(self):
    print("Gas Car")
    self.start_engine()
    self.refuel()
```

```python
print(
    "=== Vehicle System ===\n"
)

print(
    "Creating HybridCar..."
)

hybrid = HybridCar(
    battery_capacity=50,
    fuel_tank_capacity=40
)

print(
    "\nMRO for HybridCar:"
)

for cls in HybridCar.__mro__:
    print(cls.__name__)

hybrid.vehicle_info()

print(
    "\n\nCreating ElectricCar..."
)

electric = ElectricCar(
    battery_capacity=75
)

print(
    "\nMRO for ElectricCar:"
)

for cls in ElectricCar.__mro__:
    print(cls.__name__)

electric.vehicle_info()

print(
    "\n\nCreating GasCar..."
)

gas = GasCar(
    fuel_tank_capacity=55
)

print(
    "\nMRO for GasCar:"
)

for cls in GasCar.__mro__:
    print(cls.__name__)

gas.vehicle_info()
```

#Question 6: Payment System with Abstraction

```python
from abc import ABC, abstractmethod
class PaymentMethod(ABC):
    @abstractmethod
    def process_payment(self, amount):
        pass
    @abstractmethod
    def validate_payment(self):
      pass

    def get_payment_info(self):
        return (
            f"Payment Method: "
            f"{self.__class__.__name__}"
        )
```

```python
class CreditCard(PaymentMethod):
  def __init__(self, card_number, cardholder_name, expiry_date):
    self.card_number = card_number
    self.cardholder_name = cardholder_name
    self.expiry_date = expiry_date
  def validate_payment(self):
        if (len(self.card_number) == 1 and self.card_number.isdigit()):
          print("credit card validated")
          return True
        else: print("Invalid Credit Card")
        return False
  def process_payment(self, amount):
    if self.validate_payment():
      print(f"Processing credit card payment of Rs {amount}")
      print("Payment successful")
  def get_payment_info(self):
        return (
            f"Credit Card: "
            f"****{self.card_number[-4:]} "
            f"({self.cardholder_name})"
        )
```

```python
class PayPal(PaymentMethod):
  def __init__(self, email, password):
    self.email = email
    self.password = password
  def validate_payment(self):
    if ('@' not in self.email) or (self.password == ''):
        raise Exception('Invalid email or password')
  def process_payment(self, amount):
    if self.validate_payment():
      print(f"Processing PayPal payment of Rs {amount}")
      print("Payment successful")
  def get_payment_info(self):
        return (
            f"PayPal: "
            f"{self.email}"
        )
```

```python
class BankTransfer:
  def __init__(self, account_number, routing_number):
    self.account_number = account_number
    self.routing_number = routing_number
  def validate_payment(self):
    if (len(self.account_number) == 6 and self.account_number.isdigit()) and (len(self.routing_number) == 4 and self.routing_number.isdigit()):
      print( "Bank account validated.")
    else: print("Invalid Bank Account")
  def process_payment(self, amount):
    if self.validate_payment():
      print(f"Processing bank transfer payment of Rs {amount}")
      print("Payment successful")
  def get_payment_info(self):
        return (
            f"Bank Transfer: "
            f"{self.account_number} "
            f"({self.routing_number})"
        )
```

```python
print(
    "=== Payment System ===\n"
)

print(
    "Creating payment methods...\n"
)

credit_card = CreditCard(
    "1234567812341234",
    "John Doe",
    "12/28"
)

paypal = PayPal(
    "user@example.com",
    "mypassword"
)

bank = BankTransfer(
    "567890",
    "1234"
)

print(
    credit_card.get_payment_info()
)

print(
    paypal.get_payment_info()
)

print(
    bank.get_payment_info()
)

print(
    "\nProcessing payments "
    "polymorphically...\n"
)

payments = [

    (credit_card, 1000),

    (paypal, 500),

    (bank, 2000)
]

for payment_method, amount in payments:

    print(
        f"Processing Rs "
        f"{amount} via "
        f"{payment_method.__class__.__name__}..."
    )

    payment_method.process_payment(
        amount
    )

    print("-" * 40)

print(
    "\nAttempting to create "
    "PaymentMethod directly..."
)

try:

    payment = PaymentMethod()

except TypeError as e:

    print(
        f"Error: {e}"
    )
```

#Question 7: Student Management System (Complete OOP)

```python
class Person:
  def __init__(self, name, age, __email):
    self.name = name
    self.age = age
    self.__email = __email
  def get_email(self):
    return self.__email
  def set_email(self, new_email):
    if '@' not in new_email:
      raise ValueError("Invalid email address")
    self.__email = new_email
  def display_info(self):
    print(f"Name: {self.name}")
    print(f"Age: {self.age}")
    print(f"Email: {self.__email}")
```

```python
class Student(Person):
  def __init__(self, name, age, email, student_id):
    super().__init__(name, age, email)
    self.student_id = student_id
    self.__grades = {} #private dictionary

  def add_grade(
      self,
      subject,
      grade
  ):
    if not (
          0 <= grade <= 100
      ):

          print(
              f"Invalid grade "
              f"for {subject}"
          )

          return
    self.__grades[subject] = grade

  def get_grades(self, subject):
    return self.__grades.get(subject, None)

  def calculate_gpa(self):
    if not self.__grades:
      return 0
    return sum(self.__grades.values()) / len(self.__grades)

  def get_letter_grade(self):
    gpa = self.calculate_gpa()
    if gpa >= 90:
      return 'A'
    elif gpa >= 80:
      return 'B'
    elif gpa >= 70:
      return 'C'
    elif gpa >= 60:
      return 'D'
    else:
      return 'F'

  def display_info(self):
    super().display_info()
    print(f"Student ID: {self.student_id} \n Student name: {self.name} \n Age: {self.age}")
    print(f"Email: {self.get_email()} \n GPA: {self.calculate_gpa()} \n Letter grade: {self.get_letter_grade()}" )

  def get_all_grades(self):
    return self.__grades
```

```python
class Course:
  def __init__(self, course_code, course_name, credits):
     self.course_code = (course_code)
     self.course_name = (course_name)
     self.credits = credits
     self.students = []

  def enroll_student(self, student):
    if student not in self.students:
      self.students.append(student)
      print(f"Student {student.name} enrolled in {self.course_name}.")
    else:
      print(f"Student {student.name} is already enrolled in {self.course_name}.")

  def remove_student(self, student_id):
    if student_id in self.students:
      self.students.remove(student_id)
    else: print(f"Student {student_id} is not enrolled in {self.course_name}.")

  def get_course_average(self):
    if not self.students:
      return 0
    total_gpa = sum(student.calculate_gpa() for student in self.students)
    return total_gpa / len(self.students)

  def display_course_info(self):
    print(
            f"\nCourse: "
            f"{self.course_code}"
            f" - "
            f"{self.course_name}"
        )

    print(
        f"Credits: "
        f"{self.credits}"
    )

    print(
        f"Total Students: "
        f"{len(self.students)}"
    )

    print(
        f"Class Average: "
        f"{self.get_course_average():.2f}"
    )

    print(
        "\nEnrolled Students:"
    )

    for student in self.students:

        print(
            f"- "
            f"{student.name}"
        )
```

```python
print(
    "=== Student Management System ===\n"
)

print(
    "Creating students...\n"
)

student1 = Student(
    "Alice Johnson",
    20,
    "alice@university.edu",
    "S001"
)

student2 = Student(
    "Bob Smith",
    21,
    "bob@university.edu",
    "S002"
)

print(
    "Adding grades...\n"
)

student1.add_grade(
    "Math",
    95
)

student1.add_grade(
    "Science",
    88
)

student1.add_grade(
    "English",
    92
)

student2.add_grade(
    "Math",
    80
)

student2.add_grade(
    "Science",
    75
)

student2.add_grade(
    "English",
    85
)

print(
    "Displaying Student Information:\n"
)

people = [
    student1,
    student2
]

for person in people:

    person.display_info()

    print("-" * 35)

print(
    "\nCreating course...\n"
)

course = Course(
    "CS101",
    "Introduction to Programming",
    3
)

print(
    "Enrolling students...\n"
)

course.enroll_student(
    student1
)

course.enroll_student(
    student2
)

print(
    "\nCourse Statistics:"
)

course.display_course_info()

print(
    "\nEncapsulation Test:"
)

try:

    print(
        student1.__grades
    )

except AttributeError:

    print(
        "Cannot access "
        "__grades directly."
    )

try:

    print(
        student1.__email
    )

except AttributeError:

    print(
        "Cannot access "
        "__email directly."
    )
```

```python
from datetime import datetime

class Product:
  def __init__(self, product_id, name, price, stock_quantity):
    self.product_id = product_id
    self.name = name
    self.price = price
    self.__stock_quantity = stock_quantity

  def get_stock(self):
    return self.__stock_quantity
  def set_stock(self, new_quantity):
    if new_quantity >= 0:
      self.__stock_quantity = new_quantity
    else: print("Invalid stock quantity")

  def valid_stock(self):
    if self.__stock_quantity > 0:
      return True
    else: return False

  def update_stock(self, new_quantity):
    if self.__stock_quantity + new_quantity <0:
      raise ValueError(f"Invalid stock quantity for {self.name}")
    self.__stock_quantity += new_quantity
    return True

  def is_in_stock(self):
    if self.valid_stock():
      print("In stock")
    else: print("Out of stock")

  def apply_discount(self, percentage):
    if self.percentage < 0 or self.percentage > 100:
      raise ValueError("Invalid discount percentage")
    else: self.price = self.price * (1 - percentage / 100)
    return self


  def __str__(self):

      return (
          f"{self.name} "
          f"(Rs {self.price}, "
          f"Stock: {self.__stock_quantity})"
      )
```

```python
class Customer:
  def __init__(self, customer_id, name, email, address):
    self.customer_id = customer_id
    self.name = name
    self.email = email
    self.__address = address

  def get_address(self):
    return self.__address
  def set_address(self, address):
    self.__address = address

  def display_profile(self):
    print(f"Customer ID: {self.customer_id}")
    print(f"Name: {self.name}")
    print(f"Email: {self.email}")
    print(f"Address: {self.__address}")
```

```python
class Order:
  def __init__(self, order_id, customer):
    self.order_id = order_id
    self.customer = customer
    self.products = {}
    self.order_date = datetime.now()
    self.__total_amount = 0

  def add_product(self, product, quantity):
    if quantity <= 0:
      print("quantity must be positive")
      return self
    self.products[product] = quantity
    self.calculate_total()
    return self

  def remove_product(self, product_id):
    for product in list(self.products):
      if product.product_id == product_id:
        del self.products[product_id]
        self.calculate_total()
        print(f"{product.name} removed.")
        return self
    print("Product not found.")
    return self

  def calculate_total(self):
    self.__total_amount = sum(product.price * quantity for product, quantity in self.products.items())
    return self.__total_amount

  def get_total(self):
    return self.__total_amount

  def process_order(self):
    for product, quantity in self.products.items():
        if (
            product.get_stock()
            <
            quantity
        ):

            print(
                f"Not enough stock for "
                f"{product.name}"
            )

            return False

    for product, quantity in self.products.items():

        product.update_stock(
            -quantity
        )

    print(
        "Order processed successfully.")
    return True

  def display_order(self):
    print(f"Order ID: {self.order_id}")
    print(f"Customer: {self.customer.name}")
    print(f"Order Date: {self.order_date}")
    print(f"Total Amount: {self.__total_amount}")
    print("Products:")
    for product, quantity in self.products.items():
        print(f"- {product.name} (Quantity: {quantity})")
```

```python
class ShoppingCart:
  def __init__(self,customer):
    self.customer = customer
    self.items = {}

  def add_item(self,product, quantity):
    if quantity <= 0:
      print("quantity must be positive")
      return self
    self.items[product] = (self.items.get(product ,0) + quantity)
    print(
            f"Added: "
            f"{product.name} "
            f"x {quantity}"
        )
    return self

  def remove_item(self, product_id):
    for product in list(self.items):
      if product.product_id == product_id:
        del self.items[product_id]
        print(f"{product.name} removed.")
        return self
    print("Product not found.")
    return self

  def calculate_total(self):
    return sum(product.price * quantity for product, quantity in self.items.items())

  def checkout(self):
    order = Order(
            "ORD001",
            self.customer
        )

    for product, quantity in self.items.items():

        order.add_product(
            product,
            quantity
        )

    return order
```

```python
print(
    "=== E-Commerce System ===\n"
)
print(
    "Creating products...\n"
)

laptop = Product(
    "P001",
    "Laptop",
    50000,
    10
)

mouse = Product(
    "P002",
    "Mouse",
    500,
    50
)

print(laptop)
print(mouse)

print(
    "\nCreating customer...\n"
)

customer = Customer(
    "C001",
    "John Doe",
    "john@example.com",
    "Ahmedabad"
)

customer.display_profile()

print(
    "\nAdding items to cart...\n"
)

cart = ShoppingCart(
    customer
)

# Method Chaining

cart.add_item(
    laptop,
    1
).add_item(
    mouse,
    2
)

print(
    f"\nCart Total: "
    f"Rs {cart.calculate_total()}"
)

print(
    "\nProcessing checkout..."
)

order = cart.checkout()

order.process_order()

order.display_order()

print(
    "\nUpdated Stock:"
)

print(
    f"Laptop: "
    f"{laptop.get_stock()} remaining"
)

print(
    f"Mouse: "
    f"{mouse.get_stock()} remaining"
)


print(
    "\nEncapsulation Test:"
)

try:

    print(
        laptop.__stock_quantity
    )

except AttributeError:

    print(
        "Cannot access "
        "__stock_quantity directly."
    )

try:

    print(
        customer.__address
    )

except AttributeError:

    print(
        "Cannot access "
        "__address directly."
    )
```

#Question 9: Game Character System (Polymorphism & Inheritance)

```python
from abc import ABC, abstractmethod

class Character:
  def __init__(self, name, health, level):
    self.name = name
    self.health = health
    self.level = level

  @abstractmethod
  def attack(self):
    pass
  @abstractmethod
  def defend(self):
    pass
  @abstractmethod
  def use_special_ability(self):
    pass
  def is_alive(self):
    return self.health > 0
```

```python
class Warrior(Character):
  def attack(self):
    damage = 25

    print(
        f"{self.name} attacks: "
        f"Melee Strike! "
        f"(Damage: {damage})"
    )

    return damage
  def defend(self):
    print(
        f"{self.name} defends: "
        "Shield Block!"
    )

  def use_special_ability(self):
     print(
            f"{self.name} uses "
            f"Shield Bash! "
            f"(Stuns enemy)"
        )
```

```python
class Mage(Character):
  def attack(self):
    damage = 30
    print(f"{self.name} attacks: "
        f"Fireball! "
        f"(Damage: {damage}")
    return damage
  def defend(self):
    print(
        f"{self.name} casts: "
        f"Magic barrier!"
    )

  def use_special_ability(self):
    print(
        f"{self.name} uses "
        f"Fireball! "
        f"(Area Damage: 50)"
    )
```

```python
class Archer(Character):
   def attack(self):
    damage = 20
    print(
        f"{self.name} attacks: "
        f"Arrow Shot! "
        f"(Damage: {damage})"
    )
    return damage

    def defend(self):
        print(
            f"{self.name} dodges "
            f"the attack."
        )
    def use_special_ability(self):
        print(
            f"{self.name} uses "
            f"Multi-Shot! "
            f"(3 arrows, "
            f"15 damage each)"
        )
```

```python
class Game:
  def display_character_info(self, character):
    print(
          f"{character.__class__.__name__}: "
          f"{character.name} "
          f"(Health: {character.health}, "
          f"Level: {character.level})"
      )
  def battle(self, character1, character2):
    print(
            f"\n{character1.name} "
            f"vs "
            f"{character2.name}"
        )
    round_num = 1
    while (
            character1.is_alive()
            and
            character2.is_alive()
        ):

            print(
                f"\n--- Round "
                f"{round_num} ---"
            )

            damage = (
                character1.attack()
            )

            character2.health -= damage

            print(
                f"{character2.name} "
                f"Health: "
                f"{character2.health}"
            )

            if not character2.is_alive():
                print(
                    f"\nWinner: "
                    f"{character1.name}"
                )
                break

            damage = (
                character2.attack()
            )

            character1.health -= damage

            print(
                f"{character1.name} "
                f"Health: "
                f"{character1.health}"
            )

            if not character1.is_alive():
                print(
                    f"\nWinner: "
                    f"{character2.name}"
                )
                break
            round_num += 1

def perform_action(character):

    """
    Works with any object
    that implements attack()
    and use_special_ability()
    """

    character.attack()

    character.use_special_ability()
```

```python
print(
    "=== Game Character System ===\n"
)

print(
    "Creating characters...\n"
)

warrior = Warrior(
    "Conan",
    100,
    5
)

mage = Mage(
    "Gandalf",
    60,
    5
)

archer = Archer(
    "Legolas",
    80,
    5
)

game = Game()

game.display_character_info(
    warrior
)

game.display_character_info(
    mage
)

game.display_character_info(
    archer
)

print(
    "\nDemonstrating "
    "Polymorphic Behavior...\n"
)

characters = [
    warrior,
    mage,
    archer
]

for character in characters:

    character.attack()

print(
    "\nSpecial Abilities:\n"
)

for character in characters:

    character.use_special_ability()

print(
    "\nDuck Typing Demo:\n"
)

for character in characters:

    perform_action(
        character
    )

print(
    "\nBattle Simulation:"
)

game.battle(
    Warrior(
        "Conan",
        100,
        5
    ),
    Mage(
        "Gandalf",
        60,
        5
    )
)
```

#Question 10: Complete OOP Application - Bank Management System

```python
from abc import ABC, abstractmethod
from datetime import datetime
class Transaction:

    transaction_counter = 1

    def __init__(
        self,
        account_number,
        transaction_type,
        amount
    ):

        self.transaction_id = (
            f"T{Transaction.transaction_counter:03d}"
        )

        Transaction.transaction_counter += 1

        self.account_number = account_number
        self.transaction_type = transaction_type
        self.amount = amount
        self.timestamp = datetime.now()

    def __str__(self):

        return (
            f"{self.transaction_id} | "
            f"{self.transaction_type} | "
            f"Rs {self.amount} | "
            f"{self.timestamp.strftime('%Y-%m-%d %H:%M:%S')}"
        )

class Account(ABC):

    def __init__(
        self,
        account_number,
        account_holder,
        balance=0
    ):

        self.account_number = account_number
        self.account_holder = account_holder

        # Encapsulation

        self.__balance = balance

        self.transactions = []

    # Getter

    def get_balance(self):

        return self.__balance

    # Protected Setter

    def _set_balance(self, amount):

        self.__balance = amount

    @abstractmethod
    def deposit(self, amount):
        pass

    @abstractmethod
    def withdraw(self, amount):
        pass

    @abstractmethod
    def calculate_interest(self):
        pass

    def add_transaction(
        self,
        transaction_type,
        amount
    ):

        self.transactions.append(
            Transaction(
                self.account_number,
                transaction_type,
                amount
            )
        )

class SavingsAccount(Account):

    INTEREST_RATE = 0.04

    MIN_BALANCE = 1000

    def deposit(self, amount):

        if amount <= 0:
            print("Invalid amount.")
            return

        self._set_balance(
            self.get_balance() + amount
        )

        self.add_transaction(
            "Deposit",
            amount
        )

    def withdraw(self, amount):

        if (
            self.get_balance()
            - amount
            <
            self.MIN_BALANCE
        ):

            print(
                "Minimum balance violation."
            )

            return

        self._set_balance(
            self.get_balance() - amount
        )

        self.add_transaction(
            "Withdrawal",
            amount
        )

    def calculate_interest(self):

        return (
            self.get_balance()
            * self.INTEREST_RATE
        )


class CurrentAccount(Account):

    OVERDRAFT_LIMIT = 50000

    def deposit(self, amount):

        self._set_balance(
            self.get_balance() + amount
        )

        self.add_transaction(
            "Deposit",
            amount
        )

    def withdraw(self, amount):

        if (
            self.get_balance()
            - amount
            <
            -self.OVERDRAFT_LIMIT
        ):

            print(
                "Overdraft limit exceeded."
            )

            return

        self._set_balance(
            self.get_balance() - amount
        )

        self.add_transaction(
            "Withdrawal",
            amount
        )

    def calculate_interest(self):

        return 0

class FixedDeposit(Account):

    INTEREST_RATE = 0.07

    LOCK_IN_PERIOD = 1

    def deposit(self, amount):

        self._set_balance(
            self.get_balance() + amount
        )

        self.add_transaction(
            "Deposit",
            amount
        )

    def withdraw(self, amount):

        print(
            "Withdrawal not allowed "
            "before maturity."
        )

    def calculate_interest(self):

        return (
            self.get_balance()
            * self.INTEREST_RATE
        )

class Bank:

    def __init__(self, name):

        self.name = name

        self.accounts = {}

    def create_account(
        self,
        account
    ):

        self.accounts[
            account.account_number
        ] = account

        print(
            f"{account.__class__.__name__} "
            f"created: "
            f"{account.account_number}"
        )

    def close_account(
        self,
        account_number
    ):

        if account_number in self.accounts:

            del self.accounts[
                account_number
            ]

            print(
                f"{account_number} closed."
            )

    def transfer_funds(
        self,
        from_acc,
        to_acc,
        amount
    ):

        sender = self.accounts.get(
            from_acc
        )

        receiver = self.accounts.get(
            to_acc
        )

        if not sender or not receiver:

            print(
                "Account not found."
            )

            return

        old_balance = sender.get_balance()

        sender.withdraw(amount)

        if sender.get_balance() == old_balance:

            print(
                "Transfer failed."
            )

            return

        receiver.deposit(amount)

        sender.add_transaction(
            "Transfer Out",
            amount
        )

        receiver.add_transaction(
            "Transfer In",
            amount
        )

        print(
            "Transfer successful."
        )

    def generate_statement(
        self,
        account_number
    ):

        account = self.accounts.get(
            account_number
        )

        if not account:

            print(
                "Account not found."
            )

            return

        print(
            f"\nStatement for "
            f"{account_number}"
        )

        print(
            f"Holder: "
            f"{account.account_holder}"
        )

        print(
            f"Balance: "
            f"Rs {account.get_balance()}"
        )

        print(
            "\nTransactions:"
        )

        for t in account.transactions:

            print(t)


print(
    "=== Bank Management System ===\n"
)

bank = Bank(
    "OpenAI Bank"
)
print(
    "Creating accounts...\n"
)

acc1 = SavingsAccount(
    "ACC001",
    "John Doe",
    0
)

acc2 = CurrentAccount(
    "ACC002",
    "Jane Smith",
    0
)

acc3 = FixedDeposit(
    "ACC003",
    "Bob Johnson",
    10000
)

bank.create_account(acc1)
bank.create_account(acc2)
bank.create_account(acc3)

print(
    "\nProcessing transactions "
    "polymorphically...\n"
)

accounts = [
    acc1,
    acc2,
    acc3
]

acc1.deposit(10000)

print(
    f"Balance ACC001: "
    f"{acc1.get_balance()}"
)

acc1.withdraw(2000)

print(
    f"Balance ACC001: "
    f"{acc1.get_balance()}"
)

print(
    "\nCalculating interest "
    "for all accounts...\n"
)

for account in accounts:

    print(
        f"{account.account_number} "
        f"({account.__class__.__name__})"
    )

    print(
        f"Interest: Rs "
        f"{account.calculate_interest():.2f}"
    )

print(
    "\nTransferring "
    "Rs 1000 "
    "from ACC001 to ACC002..."
)

bank.transfer_funds(
    "ACC001",
    "ACC002",
    1000
)

bank.generate_statement(
    "ACC001"
)
```
