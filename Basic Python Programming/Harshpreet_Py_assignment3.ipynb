#Question 1: Exception Handling - Robust Calculator

```python
class InvalidOperationError(Exception):
    pass

def calculator(num1, num2, operation):
    try:

        if not isinstance(num1, (int, float)):
            raise TypeError("First number must be numeric.")

        if not isinstance(num2, (int, float)):
            raise TypeError("Second number must be numeric.")

        # Arithmetic Operations
        if operation == "+":
            result = num1 + num2

        elif operation == "-":
            result = num1 - num2

        elif operation == "*":
            result = num1 * num2

        elif operation == "/":
            result = num1 / num2

        else:
            raise InvalidOperationError(
                f"Operation '{operation}' is not supported."
            )

        print("Result:", result)

    except ZeroDivisionError:
        print("Error: Cannot divide by zero!")

    except TypeError as e:
        print("Error:", e)

    except OverflowError:
        print("Error: Result is too large!")

    except InvalidOperationError as e:
        print("Error: InvalidOperationError -", e)

    finally:
        print("[Finally block executed - Operation logged]")

try:

    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    operation = input(
        "Enter operation (+, -, *, /): "
    )

    calculator(num1, num2, operation)

except ValueError:
    print(
        "Error: Invalid input! "
        "Please enter a valid number."
    )
```

```python
import csv
import re

print("=== CSV File Processor ===")

input_file = "employees.csv"
output_file = "employees_valid.csv"

valid_records = []
invalid_records = 0
total_salary = 0

try:

    print(f"\nReading {input_file}...")

    with open(input_file, "r", newline="") as file:

        reader = csv.DictReader(file)

        records = list(reader)

    print(
        f"Processing {len(records)} records..."
    )

    for row_number, row in enumerate(records, start=1):

        try:
            name = row["name"].strip()
            email = row["email"].strip()

            if not name:
                print(
                    f"Row {row_number}: "
                    f"Missing name - Skipping row"
                )
                invalid_records += 1
                continue

            if not email:
                print(
                    f"Row {row_number}: "
                    f"Missing email - Skipping row"
                )
                invalid_records += 1
                continue

            email_pattern = (
                r"^[A-Za-z0-9._%+-]+"
                r"@[A-Za-z0-9.-]+"
                r"\.[A-Za-z]{2,}$"
            )

            if not re.match(
                email_pattern,
                email
            ):
                print(
                    f"Row {row_number}: "
                    f"Invalid email format - Skipping row"
                )
                invalid_records += 1
                continue
            try:
                age = int(row["age"])
            except ValueError:

                print(
                    f"Row {row_number}: "
                    f"Invalid age "
                    f"'{row['age']}' "
                    f"- Skipping row"
                )

                invalid_records += 1
                continue

            try:
                salary = float(row["salary"])
            except ValueError:

                print(
                    f"Row {row_number}: "
                    f"Invalid salary "
                    f"'{row['salary']}' "
                    f"- Skipping row"
                )

                invalid_records += 1
                continue

            valid_records.append(
                {
                    "name": name,
                    "age": age,
                    "email": email,
                    "salary": salary
                }
            )

            total_salary += salary

        except KeyError as e:

            print(
                f"Row {row_number}: "
                f"Missing column {e}"
            )

            invalid_records += 1

    valid_count = len(valid_records)

    average_salary = (
        total_salary / valid_count
        if valid_count > 0
        else 0
    )

    print("\n=== Processing Complete ===")

    print(
        f"Valid Records: {valid_count}"
    )

    print(
        f"Invalid Records: {invalid_records}"
    )

    print("\nStatistics:")

    print(
        f"Average Salary: "
        f"Rs {average_salary:,.2f}"
    )

    print(
        f"Total Employees: "
        f"{valid_count}"
    )

    print(
        f"\nWriting valid data "
        f"to '{output_file}'..."
    )

    with open(
        output_file,
        "w",
        newline=""
    ) as file:

        fieldnames = [
            "name",
            "age",
            "email",
            "salary"
        ]

        writer = csv.DictWriter(
            file,
            fieldnames=fieldnames
        )

        writer.writeheader()

        writer.writerows(valid_records)

    print(
        "File written successfully."
    )

except FileNotFoundError:

    print(
        f"Error: File "
        f"'{input_file}' not found."
    )

except PermissionError:

    print(
        "Error: Permission denied."
    )

except csv.Error as e:

    print(
        f"CSV Error: {e}"
    )

except IOError as e:

    print(
        f"I/O Error: {e}"
    )

except Exception as e:

    print(
        f"Unexpected Error: {e}"
    )
```

#Question 4: JSON Configuration Manager

```python
import json
import os

CONFIG_FILE = "config.json"

DEFAULT_CONFIG = {
    "database": {
        "host": "localhost",
        "port": 5432,
        "name": "mydb"
    },
    "api": {
        "timeout": 30,
        "retries": 3
    },
    "logging": {
        "level": "INFO",
        "file": "app.log"
    }
}

def validate_config(config):

    required_sections = [
        "database",
        "api",
        "logging"
    ]

    for section in required_sections:

        if section not in config:
            raise ValueError(
                f"Missing required section: {section}"
            )

    if not isinstance(
        config["database"]["host"],
        str
    ):
        raise TypeError(
            "Database host must be string"
        )

    if not isinstance(
        config["database"]["port"],
        int
    ):
        raise TypeError(
            "Database port must be integer"
        )


    if not (
        1 <= config["database"]["port"] <= 65535
    ):
        raise ValueError(
            "Port must be between 1 and 65535"
        )


    if not isinstance(
        config["api"]["timeout"],
        int
    ):
        raise TypeError(
            "Timeout must be integer"
        )

    if config["api"]["timeout"] <= 0:
        raise ValueError(
            "Timeout must be positive"
        )

    return True

def load_config():

    try:

        print(
            f"Loading {CONFIG_FILE}..."
        )

        with open(
            CONFIG_FILE,
            "r"
        ) as file:

            config = json.load(file)

        validate_config(config)

        print(
            "Configuration loaded successfully."
        )

        return config

    except FileNotFoundError:

        print(
            "Config file not found."
        )

        print(
            "Creating default configuration..."
        )

        save_config(DEFAULT_CONFIG)

        return DEFAULT_CONFIG

    except json.JSONDecodeError:

        print(
            "Error: Invalid JSON format "
            "in file."
        )

        print(
            "Using default configuration."
        )

        return DEFAULT_CONFIG

    except Exception as e:

        print(
            f"Configuration Error: {e}"
        )

        print(
            "Using default configuration."
        )

        return DEFAULT_CONFIG
def save_config(config):

    try:

        with open(
            CONFIG_FILE,
            "w"
        ) as file:

            json.dump(
                config,
                file,
                indent=4
            )

        print(
            f"Configuration saved to "
            f"{CONFIG_FILE}"
        )

    except IOError as e:

        print(
            f"Write Error: {e}"
        )


def display_config(config):

    print("\nCurrent Settings:")

    print(
        f"- Database Host: "
        f"{config['database']['host']}"
    )

    print(
        f"- API Timeout: "
        f"{config['api']['timeout']} seconds"
    )

    print(
        f"- Logging Level: "
        f"{config['logging']['level']}"
    )

print(
    "=== JSON Configuration Manager ==="
)

config = load_config()

display_config(config)
print(
    "\nUpdating API timeout to 60..."
)

config["api"]["timeout"] = 60

print(
    "Configuration updated."
)


print(
    "\nAdding new setting:"
    " cache.enabled = True"
)

config["cache"] = {
    "enabled": True
}

print(
    "New setting added."
)
print(
    "\nRemoving logging file setting..."
)

config["logging"].pop(
    "file",
    None
)

print(
    "Setting removed."
)
print(
    "\nSaving configuration..."
)

save_config(config)

# ------------------------------------------
# Error Handling Test
# ------------------------------------------

print(
    "\n=== Error Handling Test ==="
)

try:

    with open(
        "invalid.json",
        "w"
    ) as file:

        file.write(
            "{ invalid json"
        )

    with open(
        "invalid.json",
        "r"
    ) as file:

        json.load(file)

except json.JSONDecodeError:

    print(
        "Error: Invalid JSON format "
        "in file."
    )

    print(
        "Using default configuration."
    )
```

#Question 5: Logging System with File Management

```python
import os
from datetime import datetime

LOG_FILE = "app.log"
MAX_LOG_SIZE = 1000  # bytes

def rotate_log(log_file):

    try:

        if (
            os.path.exists(log_file)
            and os.path.getsize(log_file) > MAX_LOG_SIZE
        ):

            timestamp = datetime.now().strftime(
                "%Y-%m-%d_%H-%M-%S"
            )

            archive_name = (
                f"app_{timestamp}.log"
            )

            os.rename(
                log_file,
                archive_name
            )

            print(
                f"Log file size exceeded. Rotating..."
            )

            print(
                f"Old log archived as: "
                f"{archive_name}"
            )

    except Exception as e:

        print(
            f"Log rotation failed: {e}"
        )


def log_message(
    level,
    message,
    log_file=LOG_FILE
):

    allowed_levels = [
        "INFO",
        "WARNING",
        "ERROR",
        "DEBUG"
    ]

    if level not in allowed_levels:

        print(
            f"Invalid log level: {level}"
        )

        return

    rotate_log(log_file)

    timestamp = datetime.now().strftime(
        "%Y-%m-%d %H:%M:%S"
    )

    log_entry = (
        f"[{timestamp}] "
        f"[{level}] "
        f"{message}\n"
    )

    try:

        with open(
            log_file,
            "a"
        ) as file:

            file.write(log_entry)

        print(log_entry.strip())

    except PermissionError:

        print(
            "Permission denied."
        )

        print(
            "Fallback Console Log:"
        )

        print(log_entry.strip())

    except OSError as e:

        print(
            f"File Error: {e}"
        )

        print(
            "Fallback Console Log:"
        )

        print(log_entry.strip())

    except Exception as e:

        print(
            f"Unexpected Error: {e}"
        )

        print(
            "Fallback Console Log:"
        )

        print(log_entry.strip())


def read_recent_logs(
    log_file,
    count=5
):

    try:

        with open(
            log_file,
            "r"
        ) as file:

            lines = file.readlines()

        print(
            f"\nReading recent logs "
            f"(last {count} entries):"
        )

        for line in lines[-count:]:
            print(line.strip())

    except FileNotFoundError:

        print(
            "Log file not found."
        )


# ------------------------------------------
# Filter By Level
# ------------------------------------------

def filter_logs(
    log_file,
    level
):

    try:

        print(
            f"\nFiltering "
            f"{level} logs:"
        )

        with open(
            log_file,
            "r"
        ) as file:

            for line in file:

                if f"[{level}]" in line:
                    print(
                        line.strip()
                    )

    except FileNotFoundError:

        print(
            "Log file not found."
        )


def search_logs(
    log_file,
    keyword
):

    try:

        print(
            f"\nSearching logs "
            f"for keyword: "
            f"{keyword}"
        )

        with open(
            log_file,
            "r"
        ) as file:

            found = False

            for line in file:

                if keyword.lower() in (
                    line.lower()
                ):
                    print(
                        line.strip()
                    )
                    found = True

            if not found:

                print(
                    "No matching logs found."
                )

    except FileNotFoundError:

        print(
            "Log file not found."
        )


print(
    "=== Logging System ==="
)

print(
    "\nLogging messages..."
)

log_message(
    "INFO",
    "Application started"
)

log_message(
    "WARNING",
    "Low memory detected"
)

log_message(
    "ERROR",
    "Failed to connect to database"
)

log_message(
    "DEBUG",
    "Processing user request"
)

read_recent_logs(
    LOG_FILE
)


filter_logs(
    LOG_FILE,
    "ERROR"
)
search_logs(
    LOG_FILE,
    "database"
)

print(
    "\nLogging system demonstration complete."
)
```

#Question 6: Data Backup System

```python
import os
import shutil
import hashlib

from datetime import datetime

SOURCE_FILE = "data.txt"

with open(SOURCE_FILE, "w") as file:

    file.write(
        "This is sample data for backup.\n"
    )

    file.write(
        "Backup verification test.\n"
    )

print(
    f"Sample file '{SOURCE_FILE}' created.\n"
)

def calculate_checksum(file_path):

    md5 = hashlib.md5()

    with open(file_path, "rb") as file:

        while True:

            chunk = file.read(4096)

            if not chunk:
                break

            md5.update(chunk)

    return md5.hexdigest()

def cleanup_old_backups(
    backup_root,
    keep=10
):

    all_backups = []

    for root, dirs, files in os.walk(
        backup_root
    ):

        for file in files:

            path = os.path.join(
                root,
                file
            )

            all_backups.append(path)

    all_backups.sort(
        key=os.path.getmtime,
        reverse=True
    )

    old_files = all_backups[keep:]

    removed = 0

    for file in old_files:

        os.remove(file)

        removed += 1

    print(
        f"Kept: {min(len(all_backups), keep)} "
        f"most recent backups"
    )

    print(
        f"Removed: {removed} old backups"
    )

def backup_file(
    source_file
):

    try:

        print(
            "=== Data Backup System ==="
        )

        print(
            f"\nBacking up "
            f"'{source_file}'..."
        )


        if not os.path.exists(
            source_file
        ):

            raise FileNotFoundError(
                f"Source file "
                f"'{source_file}' "
                f"not found."
            )

        today = datetime.now().strftime(
            "%Y-%m-%d"
        )

        timestamp = datetime.now().strftime(
            "%Y-%m-%d_%H-%M-%S"
        )

        backup_dir = os.path.join(
            "backups",
            today
        )

        os.makedirs(
            backup_dir,
            exist_ok=True
        )

        base_name = os.path.splitext(
            os.path.basename(
                source_file
            )
        )[0]

        extension = os.path.splitext(
            source_file
        )[1]

        backup_file_name = (
            f"{base_name}_"
            f"{timestamp}"
            f"{extension}"
        )

        destination = os.path.join(
            backup_dir,
            backup_file_name
        )

        print(
            f"Destination: "
            f"{destination}"
        )


        shutil.copy2(
            source_file,
            destination
        )

        print(
            "Backup created successfully."
        )

        print(
            "\nVerifying backup..."
        )

        if not os.path.exists(
            destination
        ):

            raise IOError(
                "Backup verification failed."
            )

        source_size = os.path.getsize(
            source_file
        )

        backup_size = os.path.getsize(
            destination
        )

        if source_size != backup_size:

            raise IOError(
                "File size mismatch."
            )

        source_hash = calculate_checksum(
            source_file
        )

        backup_hash = calculate_checksum(
            destination
        )

        if source_hash != backup_hash:

            raise IOError(
                "Checksum mismatch."
            )

        print(
            f"Backup verified "
            f"(size: {backup_size:,} bytes)"
        )

        print(
            "\nCleaning old backups..."
        )

        cleanup_old_backups(
            "backups",
            keep=10
        )

    except FileNotFoundError as e:

        print(
            f"Error: {e}"
        )

    except PermissionError:

        print(
            "Error: Permission denied."
        )

    except OSError as e:

        print(
            f"File System Error: {e}"
        )

    except IOError as e:

        print(
            f"Verification Error: {e}"
        )

    except Exception as e:

        print(
            f"Unexpected Error: {e}"
        )


backup_file(
    SOURCE_FILE
)


print(
    "\n=== Error Test ==="
)

backup_file(
    "nonexistent.txt"
)
```

#Question 7: File Validator and Processor

```python
import os
from datetime import datetime

class InvalidFileExtensionError(Exception):
    pass


class FileSizeExceededError(Exception):
    pass


class InvalidFileFormatError(Exception):
    pass


os.makedirs("data", exist_ok=True)

# Valid TXT
with open("data/file1.txt", "w") as f:
    f.write("This is a valid text file.")

# Valid CSV
with open("data/file2.csv", "w") as f:
    f.write("id,name\n1,John")

# Invalid Extension
with open("data/file3.invalid", "w") as f:
    f.write("Invalid extension")

# Large File (for testing)
with open("data/large_file.txt", "w") as f:
    f.write("A" * 20000)      # 20 KB

print("Sample files created.\n")

DATA_DIR = "data"
ERROR_REPORT = "error_report.txt"

SUPPORTED_EXTENSIONS = {
    ".txt",
    ".csv"
}

MAX_FILE_SIZE = 10000   # 10 KB

error_log = []


def log_error(file_name, error_type, message):

    timestamp = datetime.now().strftime(
        "%Y-%m-%d %H:%M:%S"
    )

    error_log.append(
        f"[{timestamp}] "
        f"{file_name} | "
        f"{error_type} | "
        f"{message}"
    )


def validate_file(file_path):

    # File Exists

    if not os.path.exists(file_path):

        raise FileNotFoundError(
            "File does not exist."
        )

    # Read Permission

    if not os.access(file_path, os.R_OK):

        raise PermissionError(
            "Read permission denied."
        )

    # Extension Check

    extension = os.path.splitext(
        file_path
    )[1].lower()

    if extension not in SUPPORTED_EXTENSIONS:

        raise InvalidFileExtensionError(
            f"Extension '{extension}' "
            f"not supported."
        )

    # File Size Check

    size = os.path.getsize(file_path)

    if size > MAX_FILE_SIZE:

        raise FileSizeExceededError(
            f"File size ({size} bytes) "
            f"exceeds limit "
            f"({MAX_FILE_SIZE} bytes)"
        )

    # Encoding Check

    try:

        with open(
            file_path,
            "r",
            encoding="utf-8"
        ) as file:

            file.read()

    except UnicodeDecodeError:

        raise InvalidFileFormatError(
            "Invalid UTF-8 encoding."
        )

    return True

def process_file(file_path):

    with open(
        file_path,
        "r",
        encoding="utf-8"
    ) as file:

        content = file.read()

    return {
        "characters": len(content),
        "lines": len(content.splitlines())
    }


print(
    "=== File Validator and Processor ==="
)

print(
    f"\nProcessing files in "
    f"directory: ./{DATA_DIR}/"
)

total_files = 0
processed_files = 0
failed_files = 0

for file_name in os.listdir(DATA_DIR):

    file_path = os.path.join(
        DATA_DIR,
        file_name
    )

    total_files += 1

    print(
        f"\nProcessing {file_name}..."
    )

    try:

        validate_file(file_path)

        result = process_file(
            file_path
        )

        print(
            "✓ Validated and "
            "processed successfully"
        )

        print(
            f"Characters: "
            f"{result['characters']}"
        )

        print(
            f"Lines: "
            f"{result['lines']}"
        )

        processed_files += 1

    except InvalidFileExtensionError as e:

        print(
            f"✗ Error: "
            f"InvalidFileExtensionError - "
            f"{e}"
        )

        log_error(
            file_name,
            "Extension Error",
            str(e)
        )

        failed_files += 1

    except FileSizeExceededError as e:

        print(
            f"✗ Error: "
            f"FileSizeExceededError - "
            f"{e}"
        )

        log_error(
            file_name,
            "Size Error",
            str(e)
        )

        failed_files += 1

    except InvalidFileFormatError as e:

        print(
            f"✗ Error: "
            f"InvalidFileFormatError - "
            f"{e}"
        )

        log_error(
            file_name,
            "Format Error",
            str(e)
        )

        failed_files += 1

    except PermissionError as e:

        print(
            f"✗ Error: "
            f"PermissionError - "
            f"{e}"
        )

        log_error(
            file_name,
            "Permission Error",
            str(e)
        )

        failed_files += 1

    except FileNotFoundError as e:

        print(
            f"✗ Error: "
            f"FileNotFoundError - "
            f"{e}"
        )

        log_error(
            file_name,
            "File Not Found",
            str(e)
        )

        failed_files += 1

    except Exception as e:

        print(
            f"✗ Unexpected Error: "
            f"{e}"
        )

        log_error(
            file_name,
            "Unexpected Error",
            str(e)
        )

        failed_files += 1


with open(
    ERROR_REPORT,
    "w"
) as report:

    report.write(
        "=== Error Report ===\n\n"
    )

    for error in error_log:

        report.write(
            error + "\n"
        )


print(
    "\n=== Processing Summary ==="
)

print(
    f"Total Files: {total_files}"
)

print(
    f"Processed: {processed_files}"
)

print(
    f"Failed: {failed_files}"
)

print(
    f"Errors logged to: "
    f"{ERROR_REPORT}"
)
```

#Question 8: Comprehensive File Management System

```python
import os
import csv
import json
import shutil
from datetime import datetime

os.makedirs("backups", exist_ok=True)

ERROR_LOG = "error_log.txt"
OPERATION_LOG = "operations.log"

SUPPORTED_EXTENSIONS = {
    ".txt",
    ".csv",
    ".json"
}

locked_files = set()

with open("employees.csv", "w", newline="") as file:

    writer = csv.writer(file)

    writer.writerow(
        ["name", "age", "email", "salary"]
    )

    writer.writerow(
        ["John Doe", 30,
         "john@example.com", 50000]
    )

    writer.writerow(
        ["Jane Smith", 25,
         "jane@example.com", 45000]
    )

    writer.writerow(
        ["Bob Johnson", 35,
         "bob@example.com", 60000]
    )

with open("config.json", "w") as file:

    json.dump(
        {
            "app_name": "My Application",
            "version": "1.0.0",
            "settings": {
                "debug": False,
                "timeout": 30
            }
        },
        file,
        indent=4
    )

def log_error(message):

    with open(
        ERROR_LOG,
        "a"
    ) as file:

        file.write(
            f"[{datetime.now()}] "
            f"{message}\n"
        )


def log_operation(message):

    with open(
        OPERATION_LOG,
        "a"
    ) as file:

        file.write(
            f"[{datetime.now()}] "
            f"{message}\n"
        )

def validate_file(filename):

    extension = os.path.splitext(
        filename
    )[1].lower()

    if extension not in SUPPORTED_EXTENSIONS:

        raise ValueError(
            f"Unsupported format: "
            f"{extension}"
        )

def create_backup(filename):

    if not os.path.exists(filename):
        return

    timestamp = datetime.now().strftime(
        "%Y%m%d_%H%M%S"
    )

    backup_name = (
        f"backups/"
        f"{timestamp}_"
        f"{os.path.basename(filename)}"
    )

    shutil.copy2(
        filename,
        backup_name
    )

def lock_file(filename):

    if filename in locked_files:

        raise RuntimeError(
            "File is currently locked."
        )

    locked_files.add(filename)


def unlock_file(filename):

    locked_files.discard(
        filename
    )


def create_file():

    try:

        filename = input(
            "Enter filename: "
        )

        validate_file(filename)

        print(
            "Enter content "
            "(END to finish):"
        )

        lines = []

        while True:

            line = input()

            if line == "END":
                break

            lines.append(line)

        with open(
            filename,
            "w"
        ) as file:

            file.write(
                "\n".join(lines)
            )

        print(
            "File created successfully."
        )

        log_operation(
            f"Created {filename}"
        )

    except Exception as e:

        print(
            f"Error: {e}"
        )

        log_error(str(e))

def read_file():

    try:

        filename = input(
            "Enter filename: "
        )

        validate_file(filename)

        with open(
            filename,
            "r"
        ) as file:

            print(
                "\n=== File Content ==="
            )

            print(
                file.read()
            )

        log_operation(
            f"Read {filename}"
        )

    except Exception as e:

        print(
            f"Error: {e}"
        )

        log_error(str(e))

def update_file():

    filename = None

    try:

        filename = input(
            "Enter filename: "
        )

        validate_file(filename)

        lock_file(filename)

        create_backup(filename)

        print(
            "Enter new content "
            "(END to finish):"
        )

        lines = []

        while True:

            line = input()

            if line == "END":
                break

            lines.append(line)

        with open(
            filename,
            "w"
        ) as file:

            file.write(
                "\n".join(lines)
            )

        print(
            "File updated successfully."
        )

        log_operation(
            f"Updated {filename}"
        )

    except Exception as e:

        print(
            f"Error: {e}"
        )

        log_error(str(e))

    finally:

        if filename:
            unlock_file(filename)

def delete_file():

    try:

        filename = input(
            "Enter filename: "
        )

        create_backup(filename)

        os.remove(filename)

        print(
            "File deleted successfully."
        )

        log_operation(
            f"Deleted {filename}"
        )

    except Exception as e:

        print(
            f"Error: {e}"
        )

        log_error(str(e))

def list_files():

    print(
        "\n=== Files ==="
    )

    for file in os.listdir():

        if os.path.isfile(file):
            print(file)

def search_files():

    keyword = input(
        "Enter keyword: "
    )

    print(
        "\nMatching Files:"
    )

    for file in os.listdir():

        if not os.path.isfile(file):
            continue

        try:

            with open(
                file,
                "r",
                encoding="utf-8"
            ) as f:

                content = f.read()

                if keyword in content:

                    print(file)

        except:
            pass

def file_statistics():

    try:

        filename = input(
            "Enter filename: "
        )

        with open(
            filename,
            "r"
        ) as file:

            content = file.read()

        stats = os.stat(filename)

        print(
            "\n=== File Statistics ==="
        )

        print(
            f"Size: "
            f"{stats.st_size} bytes"
        )

        print(
            f"Lines: "
            f"{len(content.splitlines())}"
        )

        print(
            f"Words: "
            f"{len(content.split())}"
        )

        print(
            f"Characters: "
            f"{len(content)}"
        )

        print(
            "Created:",
            datetime.fromtimestamp(
                stats.st_ctime
            )
        )

        print(
            "Modified:",
            datetime.fromtimestamp(
                stats.st_mtime
            )
        )

    except Exception as e:

        print(
            f"Error: {e}"
        )

        log_error(str(e))

while True:

    print(
        "\n=== File Management System ==="
    )

    print("1. Create File")
    print("2. Read File")
    print("3. Update File")
    print("4. Delete File")
    print("5. List Files")
    print("6. Search Files")
    print("7. File Statistics")
    print("8. Exit")

    choice = input(
        "Enter choice: "
    )

    if choice == "1":
        create_file()

    elif choice == "2":
        read_file()

    elif choice == "3":
        update_file()

    elif choice == "4":
        delete_file()

    elif choice == "5":
        list_files()

    elif choice == "6":
        search_files()

    elif choice == "7":
        file_statistics()

    elif choice == "8":

        print(
            "Exiting..."
        )

        break

    else:

        print(
            "Invalid choice."
        )
```
